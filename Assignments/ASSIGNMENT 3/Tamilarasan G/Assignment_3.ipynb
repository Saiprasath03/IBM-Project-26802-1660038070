```python
data["Length"] = np.where(data["Length"]<0.25,0.55,data['Length']) 
data["Diameter"] = np.where(data["Diameter"]<0.2,0.4,data['Diameter']) 
data["Height"] = np.where(data["Height"]<0.05,0.15,data['Height'])
data["Height"] = np.where(data["Height"]>0.23,0.15,data['Height'])
data["Whole weight"] = np.where(data["Whole weight"]>2.0,0.7,data['Whole weight']) 
data["Shucked weight"] = np.where(data["Shucked weight"]>0.8,0.3,data['Shucked weight'])
data["Shell weight"] = np.where(data["Shell weight"]>0.6,0.21,data['Shell weight'])
data["Viscera weight"] = np.where(data["Viscera weight"]>0.4,0.18,data['Viscera weight'])
```
